```python
import seaborn as sns
df = sns.load_dataset('titanic');
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='embarked', y='fare', hue='survived', data=df)
plt.title("Fare Distribution by Port of Embarkation and Survivorship")
plt.show()
```

```python
corr_matrix = df[['age', 'fare', 'sibsp', 'parch', 'survived']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap of Variables with Survival")
plt.show()
```

```python
pivot_table = df.pivot_table(index=['survived', 'sex'],
                             columns='embarked',
                             values='fare',
                             aggfunc=['mean', 'count'],
                             fill_value=0)
print(pivot_table)
```
