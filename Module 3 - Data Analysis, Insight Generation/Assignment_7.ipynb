```python
import seaborn as sns
df = sns.load_dataset('titanic');
```

```python
df.info()
```

```python

import matplotlib.pyplot as plt


# Plot the number of survivors
sns.countplot(x='survived', data=df)
plt.title('Number of Survivors on the Titanic')
plt.xlabel('Survived (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()

# Number of people who survived
survived_count = df['survived'].sum()
survived_count
```

```python
# Plot the count of males and females
sns.countplot(x='sex', data=df)
plt.title('Gender Distribution on the Titanic')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Calculate the ratio of males to females
male_count = df[df['sex'] == 'male'].shape[0]
female_count = df[df['sex'] == 'female'].shape[0]
male_female_ratio = male_count / female_count
male_female_ratio
```

```python
# Clean the data by removing null values in 'class' column
df_clean = df.dropna(subset=['class'])

# Plot a pie chart for ticket classes
df_clean['class'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Ticket Class Distribution')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
import pandas as pd

# Plot the histogram and boxplot side by side
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 10), sharex=True)

# Histogram
df['fare'].plot.hist(bins=50, ax=ax1)
ax1.set_title('Histogram of Fare Paid')
ax1.set_xlabel('Fare')

# Boxplot
sns.boxplot(x='fare', data=df, ax=ax2)
ax2.set_title('Boxplot of Fare Paid')
ax2.set_xlabel('Fare')

plt.show()

# Calculate the number of people who paid between 0 and 100 dollars and more than 300 dollars
fare_0_to_100 = df[(df['fare'] >= 0) & (df['fare'] <= 100)].shape[0]
fare_above_300 = df[df['fare'] > 300].shape[0]

# Calculate the Interquartile Range (IQR)
IQR = df['fare'].quantile(0.75) - df['fare'].quantile(0.25)

fare_0_to_100, fare_above_300, IQR
```
