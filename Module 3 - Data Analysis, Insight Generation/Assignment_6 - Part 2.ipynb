```python
import seaborn as sns
df = sns.load_dataset('titanic');
```

```python
df.shape
```

```python
df.info()
df.isnull().sum()
```

```python
df.describe()
age_mean = df['age'].mean()
age_std = df['age'].std()
fare_mean = df['fare'].mean()
fare_std = df['fare'].std()
sex_mode = df['sex'].mode()[0]
embark_town_mode = df['embark_town'].mode()[0]
```

```python
print('age_mean: ', age_mean)
print('age_std: ', age_std)
print('fare_mean: ', fare_mean)
print('fare_std: ', fare_std)
print('sex_mode: ', sex_mode)
print('embark_town_mode: ', embark_town_mode)
```

```python
import matplotlib.pyplot as plt
sns.histplot(df['age'].dropna(), kde=False)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
infants_count = len(df[df['age'] < 3])
children_count = len(df[(df['age'] >= 5) & (df['age'] <= 10)])
```

```python
print('Infants count: ', infants_count)
print('Children count: ', children_count)
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([1, 2, 3, 4, 5])
b = np.array([2, 3, 4, 6, 8])
c = np.array([1, 5, 8, 9, 10])

plt.plot(a, c, label='a vs c', color='blue')
plt.plot(b, c, label='b vs c', color='green')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/drive/MyDrive/CABA - M3 - Data Analysis/Chapter 6/company_sales_data.csv')
profit = data['total_profit']

months = data['month_number']

plt.plot(months, profit, marker='o', color='red', linewidth=3)
plt.title('Monthly Profits')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.show()
```

```python

# Extracting sales data for the month of June (6th month)
june_face_cream_sales = data[data['month_number'] == 6]['facecream'].values[0]
june_face_wash_sales = data[data['month_number'] == 6]['facewash'].values[0]

# Data for plotting
products = ['Face Cream', 'Face Wash']
sales = [june_face_cream_sales, june_face_wash_sales]

# Creating the bar chart
plt.bar(products, sales, color=['blue', 'green'])

# Adding title and labels
plt.title('Number of units sold for face cream and face wash in the month of June')
plt.xlabel('Products')
plt.ylabel('Number of units sold')

# Display the plot
plt.show()

```
